```python
# 숙제2의 1번 과제
# 02-2.ipynb 파일을 수정하여 다른 정보를 추출하여 저장하는 코드를 작성

import gdown

gdown.download('https://bit.ly/3q9SZix', '20s_best_book.json', quiet=False)
```

```python
import pandas as pd

books_df = pd.read_json('20s_best_book.json')
books_df.head()
```

```python
# class_nm데이터 활용 카테고리 추출

categories = []
for category in books_df['class_nm']:
    if category:
        categories.append("".join(category))
    else:
        categories.append(None)
```

```python
# authors데이터 활용 저자 정보 추출

authors = []
for author_list in books_df['authors']:
    if author_list:
        authors.append("".join(author_list))
    else:
        authors.append(None)
```

```python
# bookname데이터 활용 Title로 변경 신규 데이터 프레임 생성
# new_df.head()를 호출하여 처음 5개의 행만을 확인

new_df = pd.DataFrame({'Title': books_df['bookname'],'Category': categories,'Author': authors})
new_df.head()
```

```python
# new_df를 호출하여 전체 행을 확인
new_df
```

```python
# 숙제2의 2번 과제
# 이번에는 쪽수와 함께 위에서 추출한 다른 정보를 동시에 저장하여 처리하는 코드를 작성해보세요. 
# 즉, def get_page_cnt(isbn): 함수 내에 쪽수와 다른 정보를 함께 추출, 이후 이를 활용하여 두 개의 컬럼을 추가하는 코드를 작성해보세요.
```

```python
# 쪽수 정보 추출

page_counts = []
for details in books_df['loan_count']:
    if details:
        page_counts.append(int(details))
    else:
        page_counts.append(None)
```

```python
# 쪽수 내용이 추가된 최종 컬럼
# bookname데이터 활용 Title로 변경 신규 데이터 프레임 생성
# new_df.head()를 호출하여 처음 5개의 행만을 확인

new_df = pd.DataFrame({
    'Title': books_df['bookname'],
    'Category': categories,
    'Author': authors,
    'Page Count': page_counts
})

new_df.head()
```

```python
# new_df를 호출하여 전체 행을 확인
new_df
```
